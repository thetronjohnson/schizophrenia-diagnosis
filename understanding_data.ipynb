```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import mne
```

```python
PATIENT_PATH = "dataset/schizophrenia/"
HEALTHY_PATH = "dataset/healthy/"
```

```python
patient = pd.read_csv(PATIENT_PATH + "patient_0.csv", index_col=0)
healthy = pd.read_csv(HEALTHY_PATH + "healthy0.csv", index_col=0)
```

```python
patient.head()
```

```python
healthy.head()
```

```python
EEG_channels = list(patient.index)
info = mne.create_info(ch_names = EEG_channels, sfreq = 128)
patient_raw = mne.io.RawArray(patient, info)
healthy_raw = mne.io.RawArray(healthy, info)
```

## Patient EEG data plot

```python
patient_raw.plot(duration=5, n_channels=16,color="darkblue")
print()
```

## Healthy EEG Data Plot

```python
healthy_raw.plot(duration=5, n_channels=16,color="green")
print()
```
